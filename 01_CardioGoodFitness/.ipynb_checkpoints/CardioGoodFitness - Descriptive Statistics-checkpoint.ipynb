# Cardio Good Fitness Case Study - Descriptive Statistics
The market research team at AdRight is assigned the task to identify the profile of the typical customer for each treadmill product offered by CardioGood Fitness. The market research team decides to investigate whether there are differences across the product lines with respect to customer characteristics. The team decides to collect data on individuals who purchased a treadmill at a CardioGoodFitness retail store during the prior three months. The data are stored in the CardioGoodFitness.csv file.

### The team identifies the following customer variables to study: 
  - product purchased, TM195, TM498, or TM798; 
  - gender; 
  - age, in years; 
  - education, in years; 
  - relationship status, single or partnered; 
  - annual household income ; 
  - average number of times the customer plans to use the treadmill each week; 
  - average number of miles the customer expects to walk/run each week; 
  - and self-rated fitness on an 1-to-5 scale, where 1 is poor shape and 5 is excellent shape.

### Perform descriptive analytics to create a customer profile for each CardioGood Fitness treadmill product line.

```python
# Load necessary packages

import numpy as np
import pandas as pd
```

```python
# Load the Cardio Dataset

cardio_df = pd.read_csv('Dataset/CardioGoodFitness.csv')
```

```python
cardio_df.head()
```

```python
cardio_df.describe(include='all')
```

```python
cardio_df.info()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

cardio_df.hist(figsize=(20,30))
plt.show()
```

```python
import seaborn as sns

sns.boxplot(x='Gender', y='Age', data = cardio_df)

plt.show()
```

```python
# Do Men and Women have same preferences when it comes to the fitness product they use ?
pd.crosstab(cardio_df['Product'],cardio_df['Gender'])
```

```python
pd.crosstab(cardio_df['Product'],cardio_df['MaritalStatus'])
```

```python
sns.countplot(x='Product', hue='Gender', data=cardio_df)
plt.show()
```

```python
pd.pivot_table(cardio_df, index=['Product','Gender'], columns=['MaritalStatus'])
```

```python
pd.pivot_table(cardio_df, index=['Product','Gender'], columns=['MaritalStatus'], aggfunc=len)
```

```python
sns.pairplot(cardio_df)
plt.show()
```

```python
cardio_df['Age'].std()
```

```python
cardio_df['Age'].mean()
```

```python
sns.distplot(cardio_df['Age'])
plt.show()
```

```python
cardio_df.hist(by='Gender', column='Age')
plt.show()
```

```python
cardio_df.hist(by='Gender', column='Income')
plt.show()
```

```python
cardio_df.hist(by='Gender', column='Miles')
plt.show()
```

```python
cardio_df.hist(by='Product', column='Miles', figsize=(20,30))
plt.show()
```
